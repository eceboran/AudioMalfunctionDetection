```python
import librosa
import os
import pandas as pd
```

```python
# get the parent directory
parent_dir = os.path.abspath('..')
data_dir = os.path.join(parent_dir, 'Data')
```

```python
# initialize new dict
metadata_dict = {
    'anomaly': [],
    'machine_type' : [],
    'model_type' : [],
    'length' : [],
    'filename' : [],
    'file_path' : [],
}
```

```python
# iterate over machine type folders
for machine_type in os.listdir(data_dir):
    machine_path = os.path.join(data_dir, machine_type)
    # iterate over model type folders
    for model_type in os.listdir(machine_path):
        model_path = os.path.join(machine_path, model_type)
        # iterate over category folders
        for category in os.listdir(model_path):
            category_path = os.path.join(model_path, category)
            # iterate over samples
            for filename in os.listdir(category_path):
                file_path = os.path.join(category_path, filename)
                length = librosa.get_duration(filename = file_path)
                #print(filename, category_path, category, model_type, machine_type, length)
                metadata_dict['filename'].append(filename)
                metadata_dict['file_path'].append(file_path)
                metadata_dict['machine_type'].append(machine_type)
                metadata_dict['model_type'].append(model_type)
                metadata_dict['length'].append(length)
                if category == 'normal':
                    metadata_dict['anomaly'].append(0)
                elif category == 'abnormal':
                    metadata_dict['anomaly'].append(1)
                
print(metadata_dict)
                
```

```python
metadata_df = pd.DataFrame.from_dict(metadata_dict)
```
